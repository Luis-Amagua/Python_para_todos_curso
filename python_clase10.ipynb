Operaciones con columnas

```python
df_pokedex['attack_total']=df_pokedex['attack']+df_pokedex['sp_attack']
```

```python
df_pokedex
```

```python
df_pokedex['tottal']=df_pokedex.get(['hp', 'attack', 'defense', 'sp_attack','sp_defense', 'speed']).sum(axis=1)
df_pokedex.head()
```

```python
#Apply
```

```python
df_pokedex['generation'].unique() # devueve los valores unicos
```

```python
# definimos una funcion para asociar los valores de 'generaicon'

def generaciones(numero):
  if numero==1:
    return 'kanto'
  elif numero==2:
    return 'Jonto'
  elif numero==3:
      return 'Hoenn'
  elif numero==4:
    return 'Sinnoh'
  elif numero==5:
    return 'Unova'
  elif numero==6:
    return 'Kalos'
  elif numero== 7:
    return 'Alola'
```

```python
# usamos la funcion apply 
df_pokedex['region']=df_pokedex['generation'].apply(generaciones)
df_pokedex
```

```python

#FILTROS

df_pokedex.shape
```

```python

#filtrar los pokemones legendarios
df_pokedex['is_legendary']==1 #"solo retorna valores verdaeros o falsos"
```

```python
df_pokedex.loc[df_pokedex['is_legendary']==1] # con loc se filtra, y se pone adentor la condicion
```

```python
import numpy
# usamos la funcion de numpy numpy.logical_and para tener dos filtros
df_pokedex.loc[numpy.logical_and(df_pokedex['is_legendary']==1, df_pokedex['region']=='kanto')] 
```

```python
df_pokedex.loc[numpy.logical_and(df_pokedex['is_legendary']==1, df_pokedex['region']=='Alola')] 
```

```python
# SORT
df_kanto=df_pokedex.loc[df_pokedex['region']=='kanto']
df_kanto
```

```python
# ordenen
df_kanto.sort_values('tottal') # se ordena de menor a mayor
```

```python
df_kanto.sort_values(by='tottal',ascending=False) # se ordena de MAYOR  a menor
```
